```python
#Modelos artigo F.Carvalhal 'LNEC Triaxial 

import os
import numpy as np
import matplotlib.pyplot as plt   # MATLAB plotting functions
import control as ct
from control.matlab import *  # MATLAB-like functions
import math

s=tf('s')

#Modelo
wsv1=502.65
wsv2=251.33
ws=62.83
zsv2=0.6
zs=0.2


#Modelos adicionais p/compracao com wsv1 diferentes
wsv12=350
wsv13=331.8
wsv14=200
wsv15=50

#FT
Gs=(wsv1*wsv2**2*ws)/(s*(s+wsv1)*(s**2+2*zsv2*wsv2*s+wsv2**2)*(s**2+2*zs*ws*s+ws**2))
Gs2=(wsv12*wsv2**2*ws)/(s*(s+wsv12)*(s**2+2*zsv2*wsv2*s+wsv2**2)*(s**2+2*zs*ws*s+ws**2))
Gs3=(wsv13*wsv2**2*ws)/(s*(s+wsv13)*(s**2+2*zsv2*wsv2*s+wsv2**2)*(s**2+2*zs*ws*s+ws**2))
Gs4=(wsv14*wsv2**2*ws)/(s*(s+wsv14)*(s**2+2*zsv2*wsv2*s+wsv2**2)*(s**2+2*zs*ws*s+ws**2))
Gs5=(wsv15*wsv2**2*ws)/(s*(s+wsv15)*(s**2+2*zsv2*wsv2*s+wsv2**2)*(s**2+2*zs*ws*s+ws**2))

#AFF FT
Gsa=Gs*s**2
Gsa2=Gs2*s**2
Gsa3=Gs3*s**2
Gsa4=Gs4*s**2
Gsa5=Gs5*s**2
```

```python
#Controlador AAF + PID
#Caso 1: Controlador Proporcional kp=10
kp1=10
Gsc1=kp1*Gs/(1+kp1*Gs)

#Caso 2:Controlador Proporcional kp=10 e AFF kf=0.0175
kp2=12.5
kf2=0.0175
Gsm2=Gs/(1+kf2*s**2*Gs)
Gsc2=kp2*Gsm2/(1+kp2*Gsm2)

#Caso 3:Controlador PD kp=7.5 Td=0.025
kp3=7.5
Td3=0.025
Cid3=kp3*(Td3*s**2+s)/s
Gsc3=Cid3*Gs/(1+Cid3*Gs)

#Caso 4:Controlador PD kp=30 Td=0.025 e AFF kf=0.0175
kp4=30
Td4=0.025
#kf4=0.0175
kf4=1.256
Gsm4=Gs/(1+kf4*s**2*Gs)
Cid4=kp4*(Td4*s**2+s)/s
Gsc4=Cid4*Gsm4/(1+Cid4*Gsm4)
```

```python
ct.damp(minreal(Gsc4))
```

```python
#Controlador AAF -> FT:Gsm
#AFF kf=0.02
#kff=0.02   #este valor não coloca os polos indicados no artigo-> kf=1.2565 sim
kff=1.2565
Gsm=Gs/(1+kff*s**2*Gs)
Gsm_r=minreal(Gsm)
ct.damp(Gsm_r)  #Polos de Gsm
```

```python
#Controlador AAF + PD (kp=30 Td=0.025) #o ganho kp=1880 e não kp=30
kpc=1880
Tdc=0.025
iTic=0
#FT=Gs
FT=Gsm4 #FT com controlador AAF
Cidc=kpc*(Tdc*s**2+s+iTic)/s   #Controlador PID
Gscco=minreal(Cidc/kpc*Gs) #FT direta de FT=Gs com controlador PID
Gscc_d=minreal(Cidc/kpc*FT)        #FT direta de FT=Gsm (AAF) com controlador PID
Gscc=Cidc*FT/(1+Cidc*FT)  #FT a.fechado
Gscc_r=minreal(Gscc)
ct.damp(Gscc_r)
```
